```python
import requests
import xml.dom.minidom as m
from tqdm import tqdm
import xml.etree.ElementTree as ET
from tqdm import tqdm
import json
```

```python
#request paper on Alzheimer
alz=requests.get('https://eutils.ncbi.nlm.nih.gov/entrez/eutils/esearch.fcgi?db=pubmed&term=Alzheimers+AND+2022[pdat&retmax=1000')
```

```python
doc=m.parseString(alz.text)
```

```python
idnodes_alz=doc.getElementsByTagName('Id') 
```

```python
idlist_alz=[]#PubMed IDs list for 1000 Alzheimers papers from 2022
for a in range(len(idnodes_alz)):
    ids=idnodes_alz[a].childNodes[0].wholeText
    idlist_alz.append(ids)
```

```python
len(idlist_alz)
```

```python
#request paper on cancer
canc=requests.get('https://eutils.ncbi.nlm.nih.gov/entrez/eutils/esearch.fcgi?db=pubmed&term=cancer+AND+2022[pdat&retmax=1000')
```

```python
c_doc=m.parseString(canc.text)
```

```python
idnodes_canc=c_doc.getElementsByTagName('Id') 
```

```python
idlist_canc=[]#PubMed IDs list for 1000 cancer papers from 2022
for b in range(len(idnodes_canc)):
    ids=idnodes_canc[b].childNodes[0].wholeText
    idlist_canc.append(ids)
```

```python
len(idlist_canc)
```

```python
idlist_canc
```

```python
data_info_alz={"db":"pubmed","retmode":"xml","id":idlist_alz} #define post parameters for alzheimers paper
data_info_canc={"db":"pubmed","retmode":"xml","id":idlist_canc} #define post parameters for cancer paper
```

```python
alz_info=requests.post("https://eutils.ncbi.nlm.nih.gov/entrez/eutils/efetch/fcgi",data=data_info_alz) #post request to access articles
canc_info=requests.post("https://eutils.ncbi.nlm.nih.gov/entrez/eutils/efetch/fcgi",data=data_info_canc)
```

```python
#establishing tree for alzheimers paper
```

```python
tree_alz=ET.fromstring(alz_info.text)
# alz_str=ET.tostring(tree,method='text').decode()
```

```python
len(tree_alz) #check length of tree==1000
```

```python
dictionary_list_alz=[] #dictionary list of individual dictionaries that correspond to each layer of PubmedArticle 
for index in tqdm(range(len(tree_alz))):
    article=tree_alz.findall('PubmedArticle')[index]
    title=ET.tostring(article.find('.//ArticleTitle'),method='text').decode()
    if article.find('.//AbstractText')is not None:
        abstract=[ET.tostring(node,method='text').decode() for node in article.findall(".//AbstractText")]
    else:
        abstract= ''
    query='Alzheimer'
    dictionary={"ArticleTitle":title,"Abstract":abstract,"query":query}
    dictionary_list_alz.append(dictionary)
```

```python
dictionary_alz=dict(zip(idlist_alz,dictionary_list_alz))#matching alzhemier paper id to individual dictionaries
```

```python
#establishing tree for cancer paper
```

```python
tree_canc=ET.fromstring(canc_info.text)
# alz_str=ET.tostring(tree,method='text').decode()
```

```python
len(tree_canc) #check length of tree==1000
```

```python
dictionary_list_canc=[] #dictionary list of individual dictionaries that correspond to each layer of PubmedArticle 
for index in tqdm(range(len(tree_canc))):
    article=tree_canc.findall('PubmedArticle')[index]
    title=ET.tostring(article.find('.//ArticleTitle'),method='text').decode()
    if article.find('.//AbstractText')is not None:
        abstract=[ET.tostring(node,method='text').decode() for node in article.findall(".//AbstractText")]
    else:
        abstract=''
    query='Cancer'
    dictionary={"ArticleTitle":title,"Abstract":abstract,"query":query}
    dictionary_list_canc.append(dictionary)
```

```python
dictionary_canc=dict(zip(idlist_canc,dictionary_list_canc))#matching cancer paper id to individual dictionaries
```

```python
dictionary_json={**dictionary_alz, **dictionary_canc} #merging both dictionaries for Alzheimers and cancer
```

```python
with open('./example1_1.json','w') as file: #writing json file...
    json.dump(dictionary_json,file)
```
